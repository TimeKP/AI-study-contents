```python
import pandas as pd
import numpy as np
```

```python
coef_final = pd.read_csv('/content/drive/MyDrive/AI_bootcamp/Section2-Machine Learning/Project/coef_result/coef_final')
df_20 = pd.read_csv('/content/drive/MyDrive/AI_bootcamp/Section2-Machine Learning/Project/20nm_split.csv')

df_20 = df_20.apply(pd.to_numeric, errors='coerce')
```

```python
df_20.head()
```

```python
#coef_final.drop('Unnamed: 0', axis = 1, inplace = True)
#coef_final.head()
coef_5 = coef_final[['Vg', 'Temp', 'Slope_5nm', 'Inter_5nm', 'Vds_5nm']].copy()
coef_10 = coef_final[['Vg', 'Temp', 'Slope_10nm', 'Inter_10nm', 'Vds_10nm']].copy()
coef_20 = coef_final[['Vg', 'Temp', 'Slope_20nm', 'Inter_20nm', 'Vds_20nm']].copy()
```

```python
def Extract_value(df) :
  M = 0.9
  k = 8.61733262145e-5
  epsilon_r = 7.0
  epsilon_0 = 8.8541878128e-12
  to_eV = 6.241509e18
#  permitivity = 55263499.5823

  df.columns = ['Vg', 'Temp', 'Slope', 'Inter', 'Vds']
  const = 1/(4*np.pi*epsilon_r*epsilon_0*to_eV)
  df['width'] = const/((df['Slope']*k*df['Temp']) ** 2)
  
  A = 120e-8
  M = 0.9

  df['height'] = k*df['Temp']*(np.log(A*M*(df['Temp'] ** 2)) - df['Inter'])

  return df
```

```python
coef_5 = Extract_value(coef_5)
coef_10 = Extract_value(coef_10)
coef_20 = Extract_value(coef_20)
```

```python
coef_20.head()
```

```python
df_20.head()
```

```python
column_names = ['Vds']

for Vg in range (-50, -19, 1) :
  for T in range (90, 315, 15) :
    name = str(Vg) + '/' + str(T)
    column_names.append(name)

df_20.columns = column_names

df_20.set_index('Vds', inplace = True)
df_20 = df_20.transpose()
```

```python
save_Vg_Temp = pd.DataFrame(columns = ['Vg', 'Temp'])
#print(df_5.index)
save_Vg_Temp['Vg'], save_Vg_Temp['Temp'] = df_20.index.str.split('/', 1).str
save_Vg_Temp = save_Vg_Temp.apply(pd.to_numeric)
save_Vg_Temp.head()
```

```python
df_20.reset_index(drop = True, inplace = True)
df_20.head()
```

```python
#merge_outer = pd.merge(df1,df2, how='outer',on='id')
#result3_in = pd.concat([df1,df2],axis=1, join='inner')

df_20nm = pd.concat([save_Vg_Temp, df_20], axis = 1)
df_20nm = pd.merge(df_20nm, coef_20[['Vg', 'Temp', 'Vds', 'width', 'height']], how = 'outer', on = ['Vg', 'Temp'])
```

```python
df_20nm.head()
```

```python
df_20nm = df_20nm.replace([np.inf, -np.inf], np.nan)
df_20nm = df_20nm.dropna(axis = 0)
df_20nm
```

```python
df_20nm.to_csv('20nm')
```
