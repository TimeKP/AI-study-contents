```python
import pandas as pd

coef_5 = pd.read_csv('/content/drive/MyDrive/AI_bootcamp/Section2-Machine Learning/Project/coef_result/result_5nm')
#coef_10 = pd.read_csv('/content/drive/MyDrive/AI_bootcamp/Section2-Machine Learning/Project/coef_result/result_10nm')
coef_20 = pd.read_csv('/content/drive/MyDrive/AI_bootcamp/Section2-Machine Learning/Project/coef_result/result_20nm')
```

```python
coef_5[['slope', 'inter', 'Vds']] = 0

coef_5.loc[coef_5.MAX_high == 'SE', 'slope'] = coef_5.loc[coef_5.MAX_high == 'SE', 'SE_slope_high']
coef_5.loc[coef_5.MAX_high == 'SE', 'inter'] = coef_5.loc[coef_5.MAX_high == 'SE', 'SE_inter_high']
coef_5.loc[coef_5.MAX_high == 'SE', 'Vds'] = 'high'

coef_5.loc[coef_5.MAX_low == 'SE', ['slope', 'Vds']] = coef_5.loc[coef_5.MAX_low == 'SE', 'SE_slope_low'], 'low'
coef_5.loc[coef_5.MAX_low == 'SE', 'inter'] = coef_5.loc[coef_5.MAX_low == 'SE', 'SE_inter_low']

coef_5.head()
```

```python
coef_10_test.head()
coef_10 = coef_10_test[['Vg',	'Temp',	'MAX_low', 'SE_slope_low',	'SE_inter_low',	'MAX_high', 'SE_slope_high',	'SE_inter_high']]

coef_10[['slope', 'inter', 'Vds']] = 0

coef_10.loc[coef_10.MAX_high == 'SE', 'slope'] = coef_10.loc[coef_10.MAX_high == 'SE', 'SE_slope_high']
coef_10.loc[coef_10.MAX_high == 'SE', 'inter'] = coef_10.loc[coef_10.MAX_high == 'SE', 'SE_inter_high']
coef_10.loc[coef_10.MAX_high == 'SE', 'Vds'] = 'high'

coef_10.loc[coef_10.MAX_low == 'SE', ['slope', 'Vds']] = coef_10.loc[coef_10.MAX_low == 'SE', 'SE_slope_low'], 'low'
coef_10.loc[coef_10.MAX_low == 'SE', 'inter'] = coef_10.loc[coef_10.MAX_low == 'SE', 'SE_inter_low']

coef_10.head()
```

```python
#coef_20.drop('Unnamed: 0', axis = 1, inplace = True)

coef_20[['slope', 'inter', 'Vds']] = 0

coef_20.loc[coef_20.MAX_high == 'SE', 'slope'] = coef_20.loc[coef_20.MAX_high == 'SE', 'SE_slope_high']
coef_20.loc[coef_20.MAX_high == 'SE', 'inter'] = coef_20.loc[coef_20.MAX_high == 'SE', 'SE_inter_high']
coef_20.loc[coef_20.MAX_high == 'SE', 'Vds'] = 'high'

coef_20.loc[coef_20.MAX_low == 'SE', ['slope', 'Vds']] = coef_20.loc[coef_20.MAX_low == 'SE', 'SE_slope_low'], 'low'
coef_20.loc[coef_20.MAX_low == 'SE', 'inter'] = coef_20.loc[coef_20.MAX_low == 'SE', 'SE_inter_low']

coef_20.head()
```
